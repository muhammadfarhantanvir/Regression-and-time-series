# Regression: 

```python
# Definiere die unabhängigen Variablen (X) und die abhängige Variable (y)
X = PO_Port[['Holiday_1', 'Incident_1', 
             'Weekdays_Monday','Weekdays_Saturday','Weekdays_Sunday','Weekdays_Thursday','Weekdays_Tuesday','Weekdays_Wednesday',
             'Month_2', 'Month_3', 'Month_4', 'Month_5','Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11','Month_12' ]]  
y = PO_Port['sales']

# Füge eine Konstante hinzu, um den Intercept zu berücksichtigen
X = sm.add_constant(X)

# Erstelle und fitte das Regressionsmodell
model = sm.OLS(y, X).fit()
print(model.summary())
```

# ARIMA:

```python
# Definiere die endogene (abhängige) Variable (sales) und die exogenen Variablen (is_incident, Month_k, Wochentag, is_holiday)
#endogene_variable = PO_Port['sales']
#exogene_variable = PO_Port[['is_incident', 'Month_k', 'Wochentag', 'is_holiday']]
exogene_variable= PO_Port[['Holiday_1', 'Incident_1', 
             'Weekdays_Monday','Weekdays_Saturday','Weekdays_Sunday','Weekdays_Thursday','Weekdays_Tuesday','Weekdays_Wednesday',
             'Month_2', 'Month_3', 'Month_4', 'Month_5','Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11','Month_12' ]]  
endogene_variable = PO_Port['sales']



#exogene_variable = sm.add_constant(exogene_variable)  # Füge eine Konstante für den Intercept hinzu

# Erstelle das ARIMA-Modell mit exogenen Variablen
model = sm.tsa.ARIMA(endog=endogene_variable, exog=exogene_variable, order=(1, 0, 0))  # AR-Ordnung = 1, MA-Ordnung = 0, Differenzierung = 0

# Führe das Modell auf den Daten ein
results = model.fit()

# Zeige die Ergebnisse der Modellanpassung an
print(results.summary())
```

# ARIMAX:

```python
exog= PO_Port[['Holiday_1', 'Incident_1', 
             'Weekdays_Monday','Weekdays_Saturday','Weekdays_Sunday','Weekdays_Thursday','Weekdays_Tuesday','Weekdays_Wednesday',
             'Month_2', 'Month_3', 'Month_4', 'Month_5','Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11','Month_12' ]]  
endog= PO_Port['sales']

# Create the ARIMA model with AR order = 1 and MA order = 0 and provide exogenous variables
order = (1, 0, 0)  # AR order = 1, MA order = 0
model = sm.tsa.ARIMA(endog, exog, order=order)

# Fit the model to the data
results = model.fit()

# Display the results
print(results.summary())

```

# Prognose:

```python
# Define the exogenous features for forecasting
#exogenous_features = PO_Port[['is_incident', 'Month_k', 'Wochentag', 'is_holiday']]
exogenous_features= PO_Port[['Holiday_1', 'Incident_1', 
             'Weekdays_Monday','Weekdays_Saturday','Weekdays_Sunday','Weekdays_Thursday','Weekdays_Tuesday','Weekdays_Wednesday',
             'Month_2', 'Month_3', 'Month_4', 'Month_5','Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11','Month_12' ]]  

# Generate the forecast for the entire data period
forecast = results.predict(exog=exogenous_features)

# Add the forecasted values to the DataFrame
PO_Port["Forecast_ARIMAX"] = forecast

# Print the DataFrame with the forecasted values
PO_Port.head()
```

## Verlust (monatlich, wochentlich) jenach Regression Coef:

```python
coefficients = {
    'const': 636.3626,
    'is_incident': -8.5200,
}

PO_Port['date'] = pd.to_datetime(PO_Port['date'])


# Gruppieren Sie die Daten monatlich und berechnen Sie den Verlust pro Monat
PO_Port_monthly = PO_Port.set_index('date').resample('M').sum()
PO_Port_monthly['loss_is_incident'] = PO_Port_monthly['is_incident'] * coefficients['is_incident']

# Gruppieren Sie die Daten wöchentlich und berechnen Sie den Verlust pro Woche
PO_Port_weekly = PO_Port.set_index('date').resample('W').sum()
PO_Port_weekly['loss_is_incident'] = PO_Port_weekly['is_incident'] * coefficients['is_incident']

# Drucken Sie die Ergebnisse pro Monat und pro Woche
print("Verlust pro Monat:")
print(PO_Port_monthly[['sales', 'is_incident', 'loss_is_incident']])

print("\nVerlust pro Woche:")
print(PO_Port_weekly[['sales', 'is_incident', 'loss_is_incident']])
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Annahme: Sie haben bereits die Daten in einem DataFrame mit den Spalten 'date', 'sales', 'is_incident' und anderen relevanten Features

# Schritt 1: Segmentierung der Daten nach is_incident
Port_incident_1 = PO_Port.loc[PO_Port['is_incident'] == 1]
Port_incident_0 = PO_Port.loc[PO_Port['is_incident'] == 0]

# Schritt 2: Verwenden von ARIMAX für die Prognose

# Modell für is_incident=1
X_incident_1 = Port_incident_1[['Holiday_1', 'Weekdays_Monday', 'Weekdays_Saturday', 'Weekdays_Sunday', 'Weekdays_Thursday', 'Weekdays_Tuesday', 'Weekdays_Wednesday', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11', 'Month_12']]
y_incident_1 = Port_incident_1['sales']

# Hinzufügen einer Konstante für den Intercept (falls noch nicht hinzugefügt)
#X_incident_1 = sm.add_constant(X_incident_1)

model_incident_1 = sm.tsa.ARIMA(y_incident_1, order=(1, 0, 0), exog=X_incident_1)
model_fit_incident_1 = model_incident_1.fit()

# Prognose für is_incident=1
forecast_incident_1 = model_fit_incident_1.forecast(steps=len(Port_incident_1), exog=X_incident_1)

# Modell für is_incident=0
X_incident_0 = Port_incident_0[['Holiday_1', 'Weekdays_Monday', 'Weekdays_Saturday', 'Weekdays_Sunday', 'Weekdays_Thursday', 'Weekdays_Tuesday', 'Weekdays_Wednesday', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6', 'Month_7', 'Month_8', 'Month_9', 'Month_10', 'Month_11', 'Month_12']]
y_incident_0 = Port_incident_0['sales']

# Hinzufügen einer Konstante für den Intercept (falls noch nicht hinzugefügt)
#X_incident_0 = sm.add_constant(X_incident_0)

model_incident_0 = sm.tsa.ARIMA(y_incident_0, order=(1, 0, 0), exog=X_incident_0)
model_fit_incident_0 = model_incident_0.fit()

# Prognose für is_incident=0
forecast_incident_0 = model_fit_incident_0.forecast(steps=len(Port_incident_0), exog=X_incident_0)

# Schritt 3: Hinzufügen der Prognose-Spalte zu den DataFrames Port_incident_1 und Port_incident_0
Port_incident_1['forecast'] = forecast_incident_1
Port_incident_0['forecast'] = forecast_incident_0

# Schritt 4: Merge der beiden DataFrames Port_incident_1 und Port_incident_0
forecast_combined = pd.concat([Port_incident_1, Port_incident_0], axis=0)

print(forecast_combined)
#PO_Port.head()
```
